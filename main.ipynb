```python
model_logreg = {"model": "LogisticRegression", 
                "model_args": {"penalty": "l2", 
                               "C": 1.0, 
                               "solver": "lbfgs", 
                               "max_iter": 1000}}

model_rf = {"model": "RandomForestClassifier",
            "model_args": {"n_estimators": 20,
                           "max_depth": None,
                           "random_state": 42,
                           "n_jobs": -1}}

model_tree = {"model": "DecisionTreeClassifier",
              "model_args": {"criterion": "gini",
                             "max_depth": None,
                             "min_samples_split": 2,
                             "random_state": 42}}

## –°–ï–†–ò–Ø –≠–ö–°–ü–ï–†–ò–ú–ï–ù–¢–û–í ‚Ññ1
# –ü—Ä–æ–≤–µ—Ä—è—é, –∫–∞–∫ –±—É–¥–µ—Ç –º–µ–Ω—è—Ç—å—Å—è –∫–∞—á–µ—Å—Ç–≤–æ –º–æ–¥–µ–ª–µ–π –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç —á–∏—Å–ª–∞ –Ω–∞–±–ª—é–¥–µ–Ω–∏–π

for tr_size in [0.1, 0.5, 1.]:
    update_yaml_params("process_data", {'train_size': tr_size})

    update_yaml_params("train", model_logreg)
    run_experiment(f"exp_1_trainsize{tr_size}_logreg")

    update_yaml_params("train", model_rf)
    run_experiment(f"exp_1_trainsize{tr_size}_rf")

    update_yaml_params("train", model_tree)
    run_experiment(f"exp_1_trainsize{tr_size}_tree")

```

```python
from copy import deepcopy

## –°–ï–†–ò–Ø –≠–ö–°–ü–ï–†–ò–ú–ï–ù–¢–û–í ‚Ññ2
# –ü—Ä–æ–≤–µ—Ä—è—é, –∫–∞–∫ –º–µ–Ω—è–µ—Ç—Å—è –∫–∞—á–µ—Å—Ç–≤–æ –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏–∏ –õ–∏–Ω —Ä–µ–≥–∏

model_logreg = {"model": "LogisticRegression", 
                "model_args": {"penalty": "l2", 
                               "C": 1.0, 
                               "solver": "lbfgs", 
                               "max_iter": 1000}}



for C in [0.01, 0.1, 1, 10, 100]:
    model_args = deepcopy(model_logreg)
    model_args['model_args']['C'] = C


    update_yaml_params("process_data", {'train_size': 1.})
    update_yaml_params("train", model_args)
    run_experiment(f"exp_2_C{C}_logreg")
```
